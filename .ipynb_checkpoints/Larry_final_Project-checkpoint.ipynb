## Data Exploration

```r
#import house price 
house_price = read.csv("everything.csv", header=TRUE, na.strings = "")
```

**Selected predictors for ZHVI - (home price)**

**ZHVI** = Home prices (Response)

**Mortgage_rate** = National interest rate (Qualitative)

**Unemp_rate** = National unemployment rate (Qualitative)

**CPI_Rent** = National rent metric (Qualitative)

**CPI** = Consumer Price Index (Qualitative)

**SizeRank** (Qualitative)

### Split train and test datasets

```r
# select columns needed for macro look at home prices)
macro_data <-data.frame(house_price)|>
select(SizeRank, year_month, ZHVI, Mortgage_rate, Unemp_rate, CPI_Rent, CPI)
head(macro_data, 4)

# Generate test df; only "2023"
macro_data_test <- macro_data|> filter(grepl("^2023", year_month))
dim(macro_data_test)

# Generate train df: no "2023"
macro_data_train <-macro_data|> filter(!grepl("^2023", year_month))
dim(macro_data_train)
```

### Boxplots

See distribution of data, identify potential outliers, and understand spread of variables. 

```r
par(mfrow = c(1, 5))
boxplot(macro_data$Unemp_rate, main = "Unemployment Rates", col = "lightblue")
boxplot(macro_data$ZHVI, main = "Home Prices", col= "red")
boxplot(macro_data$Mortgage_rate, main = "Mortgage Rates", col = "lightgreen")
boxplot(macro_data$CPI_Rent, main = "CPI_Rent", col= "blue" )
boxplot(macro_data$CPI, main = "CPI", col = "gold")
```

### Correlation Matrix 

Highly correlated predictors may lead to multicollinearity.

```r
# generate a matrix for corrplot function; all numerics required 
cor_mat <-macro_data|>select(ZHVI, Mortgage_rate, Unemp_rate, CPI_Rent, CPI)

# comput a correlation matrix
cor_matrix <- cor(cor_mat, use = "complete.obs", method = "pearson")

corrplot(cor_matrix, method = "circle")
```

Indicator # 1 of multicollinearity between predictors.

## Initial Model Fit 


```r
home_price <-lm(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI, data = macro_data_train)
summary(home_price)
```

## Model Assumptions and Diagnostics Checks

```r
plot(home_price)
```

```r
hist(residuals(home_price), col = "black")
```

### Findings 

Residual plots show minor signs of non-constant variance and a depatures from normality. Several outliers are also identified which may be increasing the effects of non-constant variance and non-normality. 

### Outliers Analysis 

```r
cook <- cooks.distance(home_price)
influential <- which(cook > 4 / length(cook))  # Common rule of thumb

# show corresponding observations in the orginal dataframe cited as an outlier. 
macro_data[influential, ]
```

Given 1083 observations cited as potential outliers; it is most likely that the model is not accounting for phenomena in the data set. Outliers should remain for now. Further analyis needed to explore this pattern in the data. 

### Multicollinearity Investigation 

```r
vif(home_price)
```

Given VIF guidelines (*1-5 = low multicollinearity, > 5 potenital issues, and > 10 serious multicollinearity*) predictors CPI_Rent and CPI show significant indications of multicollinearity. This is indicator # 2 of multicollinearity between CPI_Rent and CPI predictors

## Second Model Fit 

### Ridge regression 

```r
# set up matrices 

x <-model.matrix(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI, data = macro_data_train)[,-1]

y <-macro_data_train$ZHVI

# ridge regression with 
home_price_ridge <-cv.glmnet(x, y, alpha = 0, nfolds = 10, nlambda = 100)

# find optimal lambda
opt_lambda <-home_price_ridge$lambda.min

# plot 
plot(home_price_ridge)
```

```r
# ridge model with optimal lambda
home_price_ridge <-glmnet(x, y, alpha = 0, nfolds = 10, lambda = opt_lambda, nlambda = 100)
coef(home_price_ridge)
```

```r
plot(home_price_ridge, xvar = "lambda", label = TRUE,main = "Ridge Trace Plot")
abline(v = log(opt_lambda), col = "red", lty = 2)
```
