## Data Exploration

```r
#import house price 
house_price = read.csv("everything.csv", header=TRUE, na.strings = "")
```

**Selected predictors for ZHVI - (home price)**

**ZHVI** (Zillow Home Value Index) = An estimate of the typical home value within a geographic region (e.g., city, ZIP code, county). its a smoothed, seasonally adjusted measure that reflects changes in home values over time. Units: dollars (USD). (Quantitative)

**Mortgage_rate** = National interest rate (Quantitative)

**Unemp_rate** = National unemployment rate (Quantitative)

**CPI_Rent** = A subcomponent of CPI that tracks changes in residential rental prices. Important for housing-related inflation analysis.(Quantitative)

**CPI** (Consumer Price Index) = Measures average change over time in prices paid by urban consumers for a market basket of consumer goods and services.(Quantitative)

**SizeRank** = A Zillow-specific ranking that reflects the relative size or popularity of a region (e.g., ZIP code or metro area). Lower values = larger or more tracked markets (e.g., New York) might be SizeRank = 1. (Quantitative)

### Split train and test datasets

```r
# select columns needed for macro look at home prices)
macro_data <-data.frame(house_price)|>
select(SizeRank, year_month, ZHVI, Mortgage_rate, Unemp_rate, CPI_Rent, CPI)|>
mutate(Year = substr(year_month, 1,4))
head(macro_data, 4)

# Generate test df; only "2023"
macro_data_test <- macro_data|> filter(grepl("^2023", year_month))


# Generate train df: no "2023"
macro_data_train <-macro_data|> filter(!grepl("^2023", year_month))
```

### Correlation Matrix 

Highly correlated predictors may lead to multicollinearity.

```r
# generate a matrix for corrplot function; all numerics required 
cor_mat <-macro_data|>select(ZHVI, Mortgage_rate, Unemp_rate, CPI_Rent, CPI, SizeRank)

# comput a correlation matrix
cor_matrix <- cor(cor_mat, use = "complete.obs", method = "pearson")

corrplot(cor_matrix, method = "circle")
```

Indicator # 1 of multicollinearity between CPI and CPI_Rent predictors.

## Initial Model Fit 


```r
model1 <-lm(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank, data = macro_data_train)
summary(model1)
```

```r
# cross-fold validation 
train_control <- trainControl(method = "cv", number = 10)

model1_1 <- train(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank,
               data = macro_data_train,
               method = "lm", 
               trControl = train_control)
summary(model1_1)
```

## Model Assumptions and Diagnostics Checks

```r
plot(model1)
```

```r
hist(residuals(model1), col = "black")
```

### Model Statistical Significance 

```r
confint(model1, level = 0.99)
```

### Findings 

Residual plots show minor signs of non-constant variance and a depatures from normality. Several outliers are also identified which may be increasing the effects of non-constant variance and non-normality. All estimated model predictors coefficients are statistically significant at the 95% confidence level. 

### Outliers Analysis 

```r
# Calculate Cook's distances
model1_cook <- cooks.distance(model1)

# Get the number of rows in the dataset used to create the model
cook_n <- nrow(model1$model)

# Calculate threshold
cook_threshold <- 4 / cook_n

# Plot Cook's distances
plot(model1_cook, main = "Cook's Distance", ylab = "Cook's Distance", xlab = "Observation")
abline(h = cook_threshold, col = "red", lty = 2)

# Print the calculated threshold
print(paste("Cook's distance threshold:", cook_threshold))

# Identify outliers
outliers <- which(model1_cook > cook_threshold)

# Print the number of outliers
print(paste("Number of outliers:", length(outliers)))
```

Given 1122 observations cited as potential outliers; it is most likely that the model is not accounting for phenomena in the data set. Outliers should remain for now pending further analyis needed to explore this pattern in the data. 

### Multicollinearity Investigation 

```r
#VIF Threshold 
cat("VIF threshold: ", max(10, 1/(1-summary(model1)$r.squared)), "\n")
z <-data.frame(round(vif(model1),2))
z
```

Given general VIF guidelines (*1-5 = low multicollinearity, > 5 potenital issues, and > 10 serious multicollinearity*) and the calculated VIF threshold predictors CPI_Rent and CPI show significant indications of high level of multicollinearity. This serves as indicator # 2 of multicollinearity between CPI_Rent and CPI predictors within the model. 

### Box Cox Tranformation Investigation (Normality)



```r
bc <-boxCox(model1)
```

```r
# Calc lambda for boxcox
opt.lambda.bx = bc$x[which.max(bc$y)]
cat("Optimal lambda:", round(opt.lambda.bx/0.5)*0.5, end="\n")
```

```r
# Apply lambda to response column 
macro_bx <-macro_data_train|>
mutate(ZVHI_bc = (ZHVI^opt.lambda.bx - 1)/ opt.lambda.bx)
head(macro_bx)
```

```r
# Rerun model 
model1_bx <-lm(ZVHI_bc ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank, data = macro_bx)
summary(model1_bx)
```

```r
plot(model1_bx)
```

```r
hist(residuals(model1_bx), col = "black")
```

```r
# Calculate Cook's distances
model1_cook <- cooks.distance(model1)

# Get the number of rows in the dataset used to create the model
cook_n <- nrow(model1$model)

# Calculate threshold
cook_threshold <- 4 / cook_n

# Plot Cook's distances
plot(model1_cook, main = "Cook's Distance", ylab = "Cook's Distance", xlab = "Observation")
abline(h = cook_threshold, col = "red", lty = 2)

# Print the calculated threshold
print(paste("Cook's distance threshold:", cook_threshold))

# Identify outliers
outliers <- which(model1_cook > cook_threshold)

# Print the number of outliers
print(paste("Number of outliers:", length(outliers)))
```

### Variable Selection Investigation

```r
# Stepwise Forward 
model1.2 <- step(glm(ZHVI ~ 1, family = gaussian(link = "identity"), data=macro_data_train), 
                         direction = "forward", 
                         scope = formula(glm(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank,family = gaussian(link = "identity"), data=macro_data_train)), 
                         trace = 0)
summary(model1.2)
```

```r
# Stepwise Backward
model1.3 <- step(glm(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank, family = gaussian(link = "identity"), data = macro_data_train), 
                          direction = "backward",
                          scope = formula(glm(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank, family = gaussian(link = "identity"), data = macro_data_train)),
                          trace = 0)
summary(model1.3)
```

```r
cat("Model1.3 AIC =", round(AIC(model1.3),2), "\n")
cat("Model1.3 BIC =", round(BIC(model1.3),2), "\n")
```

## Ridge Regression Model Fit

```r
# set up matrices 

x <-model.matrix(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank, data = macro_data_train)[,-1]

y <-macro_data_train$ZHVI

# ridge regression 
model4_cv <-cv.glmnet(x, y, alpha = 0, nfolds = 10, nlambda = 100)

# find optimal lambda
opt_lambda <-model4_cv$lambda.min
opt_lambda

# plot 
plot(model4_cv)
```

```r
# ridge model with optimal lambda
model4 <-glmnet(x, y, alpha = 0, nfolds = 10, nlambda = 100)
coef(model4, s = opt_lambda)
```

```r
plot(model4, xvar = "lambda", label = TRUE,main = "Ridge Trace Plot")
abline(v = log(opt_lambda), col = "red", lty = 2)
```

### Model Comparision


```r
# MSPE Model 1 
predictions <- predict(model1, newdata = macro_data_test)

actuals <- macro_data_test$ZHVI
mspe <- mean((actuals - predictions)^2)
cat("Model1 MSPE = ", mspe)
```

```r
# Show prediction intervals for model # 1
pred_test <-macro_data_test[1:15,] 

predictions2 <- predict(model1, newdata = pred_test, interval = "prediction", level = 0.99)
predictions2         
```

```r
pred_bx <-predict(model1_bx, newdata = macro_data_test)
transback <- (opt.lambda.bx * pred_bx +1) ^ (1 / opt.lambda.bx)
mspe <-mean((actuals - transback)^2)
cat("Model2 MSPE = ", mspe)
```

```r
# MSPE Model 3 

# Prep test matrix 

x_test <-model.matrix(ZHVI ~ Mortgage_rate + Unemp_rate + CPI_Rent + CPI + SizeRank, data = macro_data_test)[,-1]

y_test <-macro_data_test$ZHVI

#predict on test 
predictions <- predict(model4_cv, s = "lambda.min", newx = x_test)

mspe <-mean((y_test - predictions)^2)
cat("Model3 MSPE = ", mspe)
```
